```python
#stroop
for filename in (file_list):
    print(filename)
    newfile = filename.replace('.txt', '')
    newfile = newfile.replace('-text_to', '')

    df = pd.read_csv(filename, sep='\t')
    #df.columns = ['Subject','Trial','EventType','Code','Time','TTime','Uncertainty','Duration','Uncertainty2','ReqTime','ReqDur','StimType','PairIndex']


    #filter = df['Subject'] != ('Picture') 
    #df_unpict = df[filter]
    #filter2 = df_unpict['Subject'] != ('Event Type')
    #df_clean = df_unpict[filter2]
    #df_final = df_clean.dropna(how = 'all')
    #df_final['Time'] = df_final['Time'].astype(float)
    df["TimeSec"] = df["Time"] / 10000
    first_pulse = df.loc[0,'TimeSec']
    df['TimePulse'] = df['TimeSec'] - first_pulse
    
    #filtro per congruent
    cong_filt = df.loc[(df['Condition'] == 'congruent')]

    cong_filt['TimePulse'].to_csv(f"{newfile}_cong.txt", header=None, index=None, sep='\n', mode='a')

    #filtro per incongruent
    incong_filt = df.loc[(df['Condition'] == 'incongruent')]

    incong_filt['TimePulse'].to_csv(f"{newfile}_incong.txt", header=None, index=None, sep='\n', mode='a')

    #filtro per neutral
    neutral_filt = df.loc[(df['Condition'] == 'neutral')]

    neutral_filt['TimePulse'].to_csv(f"{newfile}_neutral.txt", header=None, index=None, sep='\n', mode='a')
```
